```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('/content/Mall_Customers.csv')
# as we know that in clustering there is no dependent variable so we have to use 
# all the features but in there is dataset customer id do not affect any feature
# so we cal easily exclude them
#  we include only annual income and spending score it help to identify the pattern
x=dataset.iloc[:,[3,4]].values
```

Using the elbow method to find the optimal num of cluster

```python
from sklearn.cluster import KMeans
# we from run for for 10 differnt cluster
wcss = []
for  i in range(1,11):
  kmeans=KMeans(n_clusters= i , init='k-means++', random_state=42 )
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)
  
plt.plot(range(1, 11), wcss)
plt.title('The elbow method')
plt.xlabel('Number of cluster')
plt.ylabel('Wcss')
plt.show()

```

from number of cluster 5 the value os wcss starts decreasing almost flat

traing the kmeans of model on the dataset

```python
  kmeans=KMeans(n_clusters= 5 , init='k-means++', random_state=42 )
  y_kmeans=kmeans.fit_predict(x)
  print(y_kmeans)
```

visualizing the cluster

```python
plt.scatter(x[y_kmeans==0,0],x[y_kmeans==0,1],s=100 , c='red' , label='cluster1')
plt.scatter(x[y_kmeans==1,0],x[y_kmeans==1,1],s=100 , c='blue' , label='cluster2')
plt.scatter(x[y_kmeans==2,0],x[y_kmeans==2,1],s=100 , c='yellow' , label='cluster3')
plt.scatter(x[y_kmeans==3,0],x[y_kmeans==3,1],s=100 , c='green' , label='cluster4')
plt.scatter(x[y_kmeans==4,0],x[y_kmeans==4,1],s=100 , c='orange' , label='cluster5')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=200,c='black',label='centroid')

plt.title('Clusters of customers')
plt.xlabel('Annual income')
plt.ylabel('Wcss')
plt.legend()
plt.show()

# y_kmeans=it select all the variables for which y_kmeans=0
# x=annual income
# y=spending income
```
