```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('/content/Position_Salaries (1).csv')
x=dataset.iloc[:,1:-1]
y=dataset.iloc[:,-1]
```

```python
print(x)
print(y)
```

in this we are going to predict the previous salary of the employee to see is he saying truth or not

we're gonna skip the step where we split the data set into the training set and the test set simply because we want to leverage the maximum data in order to make our future predictionof the salary for the position level between six and seven

traiining the linear regression model on the whole dataset

```python
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(x,y)
```

train the ploynomial regression whole dataset

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=4)
x_poly=poly_reg.fit_transform(x)
lin_reg_2=LinearRegression()
lin_reg_2.fit(x_poly,y)


```

```python
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg.predict(x),color='blue')
# we take parameters x and the predicted valeu we dont takethe actual value 
plt.title("Truth or Bluff (linera regression)")
plt.xlabel('Position level')
plt.ylabel('salary')
plt.show() 
```

```python
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg_2.predict(poly_reg.fit_transform(x)),color='blue')
# we take parameters x and the predicted valeu we dont takethe actual value 
plt.title("Truth or Bluff (polynomial regression)")
plt.xlabel('Position level')
plt.ylabel('salary')
plt.show() 
```
